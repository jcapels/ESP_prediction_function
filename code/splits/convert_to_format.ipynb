```python
import pickle

def load_pickle(file_path):
    """
    Load a pickle file and return the object contained within it.

    :param file_path: Path to the pickle file.
    :return: The object loaded from the pickle file.
    """
    with open(file_path, 'rb') as file:
        return pickle.load(file)
    
def save_pickle(obj, file_path):
    """
    Save an object to a pickle file.

    :param obj: The object to be saved.
    :param file_path: Path where the pickle file will be saved.
    """
    with open(file_path, 'wb') as file:
        pickle.dump(obj, file, protocol=4)

```

```python
splits = load_pickle("splits_0_8_proteins_train_val_test.pkl")
save_pickle(splits, "splits_0_8_proteins_train_val_test_v4.pkl")

splits = load_pickle("splits_0_4_proteins_train_val_test.pkl")
save_pickle(splits, "splits_0_4_proteins_train_val_test_v4.pkl")

splits = load_pickle("splits_0_2_proteins_train_val_test_curated_changed.pkl")
save_pickle(splits, "splits_0_2_proteins_train_val_test_v4.pkl")
```
